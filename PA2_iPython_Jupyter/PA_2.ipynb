```python
import matplotlib.pyplot as plot
import numpy as nump
import collections
import re
```

```python
parseFile = open("Gettysburg-Address.txt")
line = parseFile.read().lower()
cleanLine = re.sub(r'[^\w\s]', '', line)  # Clean up the line
wordCounter = collections.Counter()
wordCounter.update(cleanLine)
wordCounter.most_common()
sortedCount = collections.OrderedDict(wordCounter.most_common(50))
```

```python
words = sortedCount.keys()
wordFreq = sortedCount.values()
idx = nump.arange(len(words))
width = 0.5

plot.figure(figsize=(20, 10))
plot.bar(idx, wordFreq, width)
font = {
    'size' : 20
}
plot.rc('font', **font)
plot.xticks(idx + width * 0.5, words, rotation='vertical')
plot.xlabel('Words')
plot.ylabel('Frequencies')
plot.show()  # Show the graph
```
